```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from concurrent.futures import ThreadPoolExecutor
import pandas as pd

BASE = "https://goteborgsalliansen.se"

session = requests.Session()
session.headers.update({"User-Agent":"Mozilla/5.0"})


def get_soup(url):
    r = session.get(url,timeout=20)
    return BeautifulSoup(r.text,"html.parser")


# ----------------------------
# PLAYERS
# ----------------------------

def get_players():

    soup = get_soup(f"{BASE}/alla-spelare/")

    players=[]

    for p in soup.select(".player-card"):

        link=p.find("a")["href"]

        players.append({
            "player_id":int(p["data-id"]),
            "name":p["data-namn"],
            "club":p["data-klubb"],
            "years":p["data-år"],
            "matches":int(p["data-matcher-total"]),
            "goals":int(p["data-mål-total"]),
            "url":urljoin(BASE,link)
        })

    return players


# ----------------------------
# PLAYER DETAILS
# ----------------------------

def scrape_player(url):

    soup=get_soup(url)

    data={"url":url}

    h1=soup.find("h1")
    if h1:
        data["name"]=h1.text.strip()

    rows=soup.select(".player-info-row")

    for r in rows:

        label=r.select_one(".label")
        value=r.select_one(".value")

        if label and value:
            data[label.text.strip()]=value.text.strip()

    return data


# ----------------------------
# MATCH LIST
# ----------------------------

def get_match_urls():

    soup=get_soup(f"{BASE}/alla-matcher/")

    urls=set()

    for a in soup.select("a[href*='/match/']"):
        urls.add(urljoin(BASE,a["href"]))

    return list(urls)


# ----------------------------
# MATCH DETAILS
# ----------------------------

def scrape_match(url):

    soup=get_soup(url)

    data={"url":url}

    h1=soup.find("h1")
    if h1:
        data["match"]=h1.text.strip()

    facts=soup.select(".match-fact")

    for f in facts:

        label=f.select_one(".label")
        value=f.select_one(".value")

        if label and value:
            data[label.text.strip()]=value.text.strip()

    return data


# ----------------------------
# RUN SCRAPER
# ----------------------------

players=get_players()

print("players:",len(players))

with ThreadPoolExecutor(10) as ex:
    player_details=list(ex.map(lambda x: scrape_player(x["url"]),players))

match_urls=get_match_urls()

print("matches:",len(match_urls))

with ThreadPoolExecutor(10) as ex:
    matches=list(ex.map(scrape_match,match_urls))


pd.DataFrame(players).to_csv("players.csv",index=False)
pd.DataFrame(player_details).to_csv("player_details.csv",index=False)
pd.DataFrame(matches).to_csv("matches.csv",index=False)

print("done")
```

```python
import requests
import time

session = requests.Session()

session.headers.update({
    "User-Agent": "Football research bot (salgo60@msn.com)"
})

WIKIDATA_ENDPOINT = "https://query.wikidata.org/sparql" 
def get_wikidata_players():

    query = """
    SELECT ?player ?playerLabel ?svwiki WHERE {
      ?player wdt:P31 wd:Q5 .
      ?player wdt:P106 wd:Q937857 .
      
      OPTIONAL {
        ?svwiki schema:about ?player ;
                schema:isPartOf <https://sv.wikipedia.org/> .
      }

      SERVICE wikibase:label {
        bd:serviceParam wikibase:language "sv,en".
      }
    }
    LIMIT 200000
    """
    query = """
        SELECT ?player ?playerLabel ?svwiki WHERE {

          ?club wdt:P31/wdt:P279* wd:Q476028 .
          ?club wdt:P17 wd:Q34 .
        
          ?player wdt:P54 ?club .
          ?player wdt:P106 wd:Q937857 .
        
          OPTIONAL {
            ?svwiki schema:about ?player ;
                    schema:isPartOf <https://sv.wikipedia.org/>.
          }
        
          SERVICE wikibase:label { bd:serviceParam wikibase:language "sv,en". }
        
        }
        """

    r = session.get(
        WIKIDATA_ENDPOINT,
        params={"query": query, "format": "json"},
        timeout=60
    )

    r.raise_for_status()

    return r.json() 
def parse_players(data):

    players=[]

    for r in data["results"]["bindings"]:

        players.append({
            "qid": r["player"]["value"].split("/")[-1],
            "name": r["playerLabel"]["value"],
            "svwiki": r.get("svwiki",{}).get("value")
        })

    return players 
import pandas as pd
from rapidfuzz import process,fuzz

wd_players = parse_players(get_wikidata_players())

wd_names = [p["name"] for p in wd_players]

def match_player(name):

    match = process.extractOne(
        name,
        wd_names,
        scorer=fuzz.token_sort_ratio
    )

    if match and match[1] > 92:

        idx = wd_names.index(match[0])

        return wd_players[idx]

    return None 

players = pd.read_csv("players.csv")

results=[]

for _,row in players.iterrows():

    m = match_player(row["name"])

    if m:

        results.append({
            "name": row["name"],
            "wikidata": m["qid"],
            "svwiki": m["svwiki"]
        })

    else:

        results.append({
            "name": row["name"],
            "wikidata": None,
            "svwiki": None
        })

df = pd.DataFrame(results)

df.to_csv("players_svwiki.csv",index=False)
print("Skapat players_svwiki.csv")
```
