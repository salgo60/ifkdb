```python
import requests

S = requests.Session()
S.headers.update({
    "User-Agent": "wikicat/1.0 (https://github.com/salgo60/ifkdb; contact: salgo60@msn.com)"
})

API_URL = "https://sv.wikipedia.org/w/api.php"
def get_category_members_with_qid(category_title):
    session = requests.Session()
    members = []
    cmcontinue = None

    while True:
        params = {
            "action": "query",
            "generator": "categorymembers",
            "gcmtitle": category_title,
            "gcmlimit": "max",
            "prop": "pageprops",
            "format": "json",
            "gcmnamespace": 0
        }

        if cmcontinue:
            params["gcmcontinue"] = cmcontinue

        response = S.get(API_URL, params=params)

        print("Status:", response.status_code)

        if response.status_code != 200:
            print(response.text)
            response.raise_for_status()

        try:
            data = response.json()
        except ValueError:
            print("Response was not JSON:")
            print(response.text[:500])
            raise

        if "query" in data:
            for page in data["query"]["pages"].values():
                members.append({
                    "title": page["title"],
                    "pageid": page["pageid"],
                    "qid": page.get("pageprops", {}).get("wikibase_item")
                })

        if "continue" in data:
            cmcontinue = data["continue"]["gcmcontinue"]
        else:
            break

    return members
```

```python
category = "Kategori:Fotbollsspelare_i_AFC_United"
players = get_category_members_with_qid(category)

for p in players[:10]:
    print(p)
```
